```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def fungsi(x):
    return (np.exp(-x)*(x**3))
a=1
b=7
n=10
h = (b-a)/(n)
#__Membuat array dengan nilai dalam interval
x = np.linspace(a,b,n)

sumasi = 0
for i in range(1, n-1):
    sumasi += fungsi(x[i])
    
hasil = 0.5*h*(fungsi(x[0]) + 2*sumasi + fungsi(x[-1]))
print(hasil)
```

```python
res = 300
graf = np.linspace(a,b,res)

plt.figure(figsize=(15,10))
plt.plot(graf,fungsi(graf))

for i in range(n):
    plt.bar(x[i], fungsi(x[i]),
           align='center',
           width=10**-5,
           edgecolor='white')
    
plt.fill_between(x, fungsi(x),
                color='black')

plt.show()
```
